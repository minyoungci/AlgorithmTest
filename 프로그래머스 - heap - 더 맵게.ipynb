```python
## heap library 활용 

import heapq

def solution(scoville, K):
    answer = 0
    heapq.heapify(scoville)      # 리스트를 힙구조로 변환 
    
    while scoville[0] < K :
        mix = heapq.heappop(scoville) + (heapq.heappop(scoville) * 2) # heappop 은 힙의 불변성 유지하면서 가장 작은 요소 반환
        heapq.heappush(scoville,mix) # heappush 는 힙 불변성을 유지하면서 item 값을 heap 으로 푸쉬
        answer += 1 
        if len(scoville) == 1 and scovile[0] < K:
            return -1
        

    return answer
    
    
```

heap 모듈 사용법

heapq.heappush(heap, item) : item을 heap에 추가
heapq.heappop(heap) : heap에서 가장 작은 원소를 pop & 리턴. 비어 있는 경우 IndexError가 호출됨. 
heapq.heapify(x) : 리스트 x를 즉각적으로 heap으로 변환함 (in linear time, O(N) )



heap 을 사용하면 배열에 데이터를 넣고 최대값과 최소값을 찾는 시간이 많이 줄어듦. 
heap 과 이진트리의 차이점은 heap은 각 노드의 값이 자식노드보다 크거나 같음. 

### 이진 트리는 탐색을 위한 구조이고 heap 은 최대/최소를 찾기 위한 구조.
