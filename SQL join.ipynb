JOIN - 두 테이블의 데이터를 일정한 조건에 의해 연결하여 마치 하나의 테이블처럼 만드는 것.

INNER  JOIN - 

LEFT OUTER JOIN - 왼쪽에 오는 테이블 기준으로 오른쪽에 있는 테이블과 비교하여 조건에 맞는 값이 있으면 JOIN 해서 가져오고 없으면 null 값을 표시합니다. 즉, 왼쪽에 있는 데이터는 무조건 가져오며 오른쪽에 오는 테이블과 JOIN을 수행하여 조건에 맞는 데이터가 없을 시 null.
