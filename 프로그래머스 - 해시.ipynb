# 해시 

해시(hash)란 데이터의 효율적 관리를 목적으로 임의의 길이의 데이터를 고정된 길이의 데이터로 매핑하는 자료구조이다.
이 때 매핑 전 원래 데이터의 값을 키(key), 매핑 후 데이터의 값을 해시값(hash value), 매핑하는 과정 자체를 해싱(hashing)라고 한다.

데이터를 빠르게 넣거나 가져올 때 사용하는 방법.

### 해시로 풀어야 하는 코딩 테스트 문제의 경우 , 리스트로 풀게되면 효율성 테스트 실패.

위의 코드는 정확성 : 66.7 % 효율성 : 12.5 % 로 효율성이 낮게 나옴.

### 1. Hash Map 만들기 

Hash Map이란 Key-Value 의 Pair를 관리하는 클래스이다. 
Key는 phone_number / Value 는 1로 설정한다. 
기본적으로 Hash map 은 위아 같이 만들어주는게 정석이다. 

    hash_map = {}                           #1 Hash map 을 만든다 
    for phone_number in phone_book:
        hash_map[phone_number] = 1
        
이 과정으로 모든 전화번호를 Hashing 합니다. = Hash map 에 추가하기 

##### Value = 1 의 의미는 숫자가 1개 존재한다는 것.

### 2. 접두어가 Hash map 에 존재하는지 찾아야함.

존재하는 모든 전화번호가 Hash map 에 등록됌.

>> if jubdoo in hash_map 

String 이라는 Key 가 현재 hash map 에 존재하는지 확인하는 코드. 

```python
# 해시맵을 이용한 풀이 

def solution(phone_book): 
    hash_map = {}                           #1 Hash map 을 만든다 
    for phone_number in phone_book:
        hash_map[phone_number] = 1
        
    for phone_number in phone_book:         #2 접두어가 Hash map 에 존재하는지 찾는다. 
        jubdoo = ""
        
        for number in phone_number:
            jubdoo += number                 #3 접두어를 찾아야 한다. ( 기존 번호와 같은 경우 제외 )
            
            if jubdoo in hash_map and jubdoo != phone_number:
                return False
    return True

print(solution(["123","1235","29289"]))
```

```python
# 다른 사람의 풀이 

def solution(phoneBook):
    phoneBook = sorted(phoneBook)

    for p1, p2 in zip(phoneBook, phoneBook[1:]):
        if p2.startswith(p1):
            return False
    return True
```

#### sorted() , sort()

list.sort() 는 숫자리스트의 경우 작은 수부터 , 문자열의 경우 알파벳 순서, 대소문자가 섞여있을 경우 대문자부터 정렬됩니다.
sort() 에서 reverse=True 는 내림차순으로 정렬, 디폴트는 False 오름차순.

sorted(정렬할 데이터) 
sorted(정렬할 데이터, reverse 파라미터 )
sorted(정렬할 데이터, key 파라미터 )
sorted(정렬할 데이터, key 파라미터, reverse 파라미터)
sorted(딕셔너리.items(), key=lambda x:x[1] ) 이를 이용해서 key= 를 x로 세팅해주어서 정렬하도록 할 수 있음.
